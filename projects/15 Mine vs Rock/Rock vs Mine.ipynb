```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score


import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

### Data loading and data preprocessing 

```python
df = pd.read_csv('sonar data.csv', header = None)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
X = df.drop(columns = 60, axis=1)
y = df[60]
```

### Spliting data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 2023)
```

## Using sklearn lib

### Model training -- logistics regression 

```python
model = LogisticRegression().fit(X_train, y_train)
```

### Model evaluation

```python
# Model evaluation on train data
train_pred = model.predict(X_train)
print('Accuracy score on train data: ', accuracy_score(train_pred, y_train))
```

```python
# Model evaluation on test data
test_pred = model.predict(X_test)
print('Accuracy score on test data: ', accuracy_score(test_pred, y_test))
```

## Using tensorflow

```python
def custom_label(x):
    if x=='R': return 1
    else: return 0
y_train = y_train.apply(lambda x : custom_label(x))
y_test = y_test.apply(lambda x : custom_label(x))
```

```python
X_train = tf.convert_to_tensor(X_train)
y_train = tf.convert_to_tensor(y_train)

X_test = tf.convert_to_tensor(X_test)
y_test = tf.convert_to_tensor(y_test)
```

```python
model = tf.keras.Sequential([
    layers.Dense(units = 32, activation = 'relu', input_shape = [60]),
    layers.Dense(units=1,activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```
