```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('movies.csv', index_col=0)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
df1 = df[features]
df1 = df1.fillna('')
```

```python
df2 = df['title']
```

```python
combined_features = df1[features[0]] + ' ' + df1[features[1]] + ' ' + df1[features[2]] + ' ' + df1[features[3]] + ' ' + df1[features[4]]
```

```python
# converting the text data to feature vectors
feature_vectors = TfidfVectorizer().fit_transform(combined_features)
```

```python
lst = list(df['title'])
```

```python
# getting the movie name from the user
fav_movies = input('Enter your favourite movie name :')
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(fav_movies, lst)
find_close_match
```

```python
idx = df1.index[df['title'] == find_close_match[0]][0]
```

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[idx]))
```

```python
sorted_recommend = sorted(similarity_score, key= lambda x : x[1], reverse = True)
```

```python
recommended_movies = [df['title'].iloc[sorted_recommend[i][0]] for i in range(len(sorted_recommend))]
recommended_df = pd.DataFrame({'recommended_movies':recommended_movies})
recommended_df.head(20)
```
